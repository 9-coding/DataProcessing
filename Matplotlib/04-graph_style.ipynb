# Graph Style

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]
```

## 그래프 영역 채우기

### `fill_between()`
- 구간 내 그래프 아래(두 그래프 사이 위아래) 면적을 채움

<br>parameters<br>

`plt.fill_between(x_range, y_range, alpha)`
- `x_range`: x축 범위
- `y_range`: y축 범위
- `alpha`: 투명도

```python
plt.plot(x, y)
plt.fill_between(x[1:3], y[1:3], alpha=0.5)
plt.show()
```

### 두 그래프 사이

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.fill_between(x[1:3], y[1:3], y2[1:3], color='lightgray', alpha=0.5)
plt.show()
```

### `fill_betweenx()`
- 구간 내 그래프 왼쪽(두 그래프 사이 양옆) 면적을 채움

<br>parameters<br>

`plt.fill_betweenx(y_range,x_range, alpha)`
- `y_range`: y축 범위
- `x_range`: x축 범위
- `alpha`: 투명도

*y, x인 것에 유의

```python
plt.plot(x, y)
plt.fill_betweenx(y[1:3], x[1:3], alpha=0.5)
plt.show()
```

### `fill()`
- 다각형 영역 채우기.
- x, y 점으로 정의되는 다각형 영역을 지정하여 채울 수 있음.

<br>parameters<br>
`plt.fill(x_range, y_range, alpha)`
- `x_range`: x축 범위
- `y_range`: y축 범위
- `alpha`: 투명도

```python
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5)
plt.show()
```

# 축 스케일 지정
기본적으로 linear 스케일로 표시되지만 log, symlog, logit으로 변경할 수 있음.


## x축 스케일 지정
`plt.xscale('')`
- linear
- log
- symlog - symmetrical log scale
- logit

```python
x = np.linspace(-10, 10, 100)
y = x ** 3
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xscale('symlog')
plt.show()
```

## y축 스케일 지정
`plt.yscale('')`
- linear
- log
- symlog - symmetrical log scale
- logit

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.yscale('log')
plt.show()
```
