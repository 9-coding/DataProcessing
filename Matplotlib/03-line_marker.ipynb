# Line and Marker

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line

## 선 종류 지정

### 포맷 문자열 사용

```python
x = [1,2,3]
y = []
for i in range(4):
  y.append(np.full((3,), i+1))

plt.plot(x, y[3], '-', label='Solid')
plt.plot(x, y[2], '--', label='Dashed')
plt.plot(x, y[1], ':', label='Dotted')
plt.plot(x, y[0], '-.', label='Dash-dot')
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(ncol=4)
plt.show()
```

### `linestyle` 지정

```python
plt.plot(x, y[3], linestyle='solid', label='Solid')
plt.plot(x, y[2], linestyle='dashed', label='Dashed')
plt.plot(x, y[1], linestyle='dotted', label='Dotted')
plt.plot(x, y[0], linestyle='dashdot', label='Dash-dot')
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(ncol=4)
plt.show()
```

### 튜플 사용
(선, 공백)과 같이 사용<br>
ex)
- (1, 1): 선과 공백이 1:1 비율 -> 점선 . . .
- (5, 1): 선이 공백보다 긺. -> dashed - - -


```python
plt.plot(x, y[3], linestyle=(0, (1, 1)), label='(0, (1, 1))') # dotted
plt.plot(x, y[2], linestyle=(0, (1, 5)), label='(0, (1, 5))') # dotted
plt.plot(x, y[1], linestyle=(0, (5, 1)), label='(0, (5, 1))') # dashed
plt.plot(x, y[0], linestyle=(0, (3, 5, 1, 5)), label='(0, (3, 5, 1, 5))') # dash-dot
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(ncol=2)
plt.show()
```

### 선 끝 모양 지정
- `butt`: 각진 끝 모양
- `round`: 둥근 끝 모양

```python
plt.plot(x, y[3], linestyle='solid', linewidth=10,
      solid_capstyle='butt', label='solid+butt')
plt.plot(x, y[2], linestyle='solid', linewidth=10,
      solid_capstyle='round', label='solid+round')

plt.plot(x, y[1], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', label='dashed+butt')
plt.plot(x, y[0], linestyle='dashed', linewidth=10,
      dash_capstyle='round', label='dashed+round')


plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.show()
```

## 선 두께 지정
- linewidth 사용

```python
plt.plot(x, y[3], linewidth=10, label='10')
plt.plot(x, y[2], linewidth=5, label='5')
plt.plot(x, y[1], linewidth=3, label='3')
plt.plot(x, y[0], linewidth=1, label='1')
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(ncol=4)
plt.show()
```

## 선 색상 지정
`color=`를 사용해서 색 지정.<br>
색을 지정하지 않으면 아래 10개 색상이 반복적으로 표시됨.


```python
x = [1,2,3]
for i in range(10):
  plt.plot(x, np.full((3,), 9-i))

plt.show()
```

기본 색상
- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

```python
x = [1,2,3]
color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
for i in range(8):
  plt.plot(x, np.full((3,), 7-i), color=color[i])

plt.show()
```

## Hex code 사용
16진수 색상코드를 사용하여 색을 지정할 수 있음.

```python
x = np.arange(5)
y = [1, 2, 4, 7, 15]

plt.plot(x, y, color='#e35f62')
plt.show()
```

# Marker
- `marker=''`를 사용하면 선에 marker 추가.
- `''`만 사용하면 선 없이 marker만 plot.

### 대표적 markers
- `'.'`: point marker
- `','`: pixel marker
- `'o'`: circle marker
- `'^'`: triangle_up marker
- `'v'`: triangle_down marker
- `'*'`: star marker
- `'+'`: plus marker
- `'x'`: x marker




```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
markers = ['.', ',', 'o', '^', 'v', '*', '+', 'x']
i = 1

plt.figure(figsize=(10, 5))
for m in markers:
  plt.subplot(2, 4, i)
  plt.plot(x, y, m)
  i += 1

plt.tight_layout()
plt.show()
```

## Marker 색 지정

`''` 안에 색, 마커 순으로 입력<br>
ex) `'bo'`: blue circle marker<br>

기본 색상
- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

```python
markers = ['b.', 'g,', 'ro', 'c^', 'mv', 'y*', 'k+', 'wx']
i = 1

plt.figure(figsize=(10, 5))
for m in markers:
  plt.subplot(2, 4, i)
  plt.plot(x, y, m)
  i += 1

plt.tight_layout()
plt.show()
```
