# Labels
title / label / legend
- `plt.title()`
- `plt.xlabel()`
- `plt.ylabel()`
- `plt.legend()`

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 그래프 제목 설정

```python
x = np.arange(6)
y = 10*np.random.rand(6)

plt.plot(x,y)
plt.title('Graph')
plt.show()
```

### 글자 형태 지정

```python
font = {'family':'serif',
         'color':'darkred',
         'size':14}

plt.plot(x,y)
plt.title('Graph', fontdict=font)
plt.show()
```

### 타이틀 위치
left / center / right

```python
plt.plot(x,y)
plt.title('Graph', loc='left')
plt.show()
```

## 축 Label 설정

### `plt.xlabel`
`plt.xlabel('X-axis', labelpad='', fontdict='', loc='')`<br>
x축의 label 설정.

### `plt.ylabel`
`plt.ylabel('Y-axis', labelpad='', fontdict='', loc='')`<br>
y축의 label 설정.

###  

- `labelpad`: 축과 label의 여백 지정.
- `fontdict`: 폰트 지정.
- `loc`: 위치 지정.


```python
plt.plot(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

### 여백 지정
- 여백(padding) 지정
- 단위는 pt.

```python
plt.plot(x,y)
plt.xlabel('X-Label', labelpad=30)
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

### 폰트 지정
`fontdict = {...}`
- 딕셔너리 형태로 지정.
- `family`: 서체
- `color`: 글자 색
- `weight`: 글자 두께
- `size`: 글자 크기


```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot(x,y)
plt.xlabel('X-Axis', fontdict=font1)
plt.ylabel('Y-Axis', fontdict=font2)
plt.show()
```

### 위치 지정
축 label의 위치를 지정.<br> default: center
- `xlabel()`: left / center / right
- `ylabel()`: bottom / centor / top

```python
plt.plot(x,y)
plt.xlabel('X-Axis', loc='right')
plt.ylabel('Y-Axis', loc='top')
plt.show()
```

### 범례(legend) 지정
- 그래프에 사용된 데이터의 정보를 전달함.
- `plt.legend()`
- `plt.plot()` 함수에 `label` 인자를 사용하여 적고 `plt.legend()`로 표시한다.


```python
y1 = np.array([5, 1, 2, 10, 7, 9])
y2 = np.array([3, 7, 4, 2, 9, 6])
y3 = np.array([4, 6, 8, 6, 4, 5])

plt.plot(y1, label='Python')
plt.plot(y2, label='ML')
plt.plot(y3, label='DL')
plt.legend()

plt.show()
```

### 위치 변경
위아래 / 좌우
- 0: best
- 1: upper right
- 2: upper left
- 3: lower left
- 4: lower right
- 5: right
- 6: center left
- 7: center right
- 8: lower center
- 9: upper center
- 10: center

```python
y1 = np.array([5, 1, 2, 10, 7, 9])
y2 = np.array([3, 7, 4, 2, 9, 6])
y3 = np.array([4, 6, 8, 6, 4, 5])

plt.plot(y1, label='Python')
plt.plot(y2, label='ML')
plt.plot(y3, label='DL')
plt.legend(loc='lower right')
# plt.legend(loc=4)와 같음.

plt.show()
```
