# 기본 그래프
### `plt.plot()`
2차원 평면의 좌표가 주어졌을 때 그 점들을 잇는 선분을 그림. <br>
`plt.plot(x='',y='',marker='',linestyle='',color='',label='')`
- x: x값 설정
- y: y값 설정
- marker (optional): 데이터 포인트를 표시에 사용
- linestyle (optional): 선 스타일 지정
- color (optional): 선 색상 지정
- label (optional): 그래프 이름 지정


### `plt.show()`
figure를 출력함과 동시에 초기화함.
- jupyter notebook에서는 입력하지 않아도 출력 가능.
- 타 IDE에서는 `plt.show()`를 입력해야 출력됨.
- 속성 관련 출력 메시지 등이 생략됨.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Linear Graph

```python
x = np.arange(1,10)
y = x*5
plt.plot(x,y)
plt.show()
```

## 꺾은선 그래프

```python
x = np.array([0, 2, 4, 6])
y = np.array([0, 120, 60, 200])

plt.plot(x, y)
plt.show()
```

## sine graph

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

## Bar Chart
### `plt.bar()`
`plt.bar(x, height, width=0.8, bottom=None, align='center', data=None, **kwargs)`
- x: x값 설정
- height: 막대의 높이
- width: 막대의 너비. 0.8이 default.


```python
x = np.arange(5)
y = [23, 16, 36, 29, 12]

plt.bar(x,y)
plt.show()
```

## Histogram

```python
a = 100 * np.random.rand(30)

plt.hist(a, bins=[0, 25, 50, 75, 100])
plt.xticks([0, 25, 50, 75, 100])
plt.show()
```

## Pie Chart

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students, labels = langs, autopct='%1.2f%%')
plt.show()
```

## Scatter Plot

```python
x = 100 * np.random.rand(30)
y = 100 * np.random.rand(30)

plt.scatter(x,y)
plt.show()
```

## Box Plot

```python
dataSet1 = np.random.normal(100, 30, 200) # mean=100 / var=30 / size=200
dataSet2 = np.random.normal(80, 20, 200)
plotData = [dataSet1, dataSet2]

plt.boxplot(plotData)
plt.show()
```
